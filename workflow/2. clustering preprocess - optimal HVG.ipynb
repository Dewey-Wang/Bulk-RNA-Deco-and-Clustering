```python
# file: scripts/biomarker_sweep_lgbm.py
# -*- coding: utf-8 -*-
import numpy as np
import pandas as pd
import scanpy as sc
from typing import Dict, List, Tuple
from sklearn.model_selection import GroupKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, balanced_accuracy_score, f1_score, adjusted_rand_score, v_measure_score
import lightgbm as lgb
import matplotlib.pyplot as plt

# ---------------- Config ----------------
AD_PATH       = "../train_data/train_adata.h5ad"
LABEL_COL     = "highLevelType"
SAMPLE_COL    = "Sample"
CT_COL        = "highLevelType"

USE_COMBAT    = True     # pseudo-bulk 後以 Sample 當 batch 校正
TOPK_LIST     = [ 60, 80, 100, 150, 200]
STRATEGIES    = ["logFC_mean", "logFC_weighted_median", "logFC_mean_fdr"]  # 要比較的選法

N_SPLITS      = 5
SEED          = 42
EARLY_STOP    = 50

LGB_PARAMS = dict(
    objective="multiclass",
    learning_rate=0.05,
    num_leaves=63,
    min_data_in_leaf=20,
    n_estimators=5000,
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=SEED,
    n_jobs=-1,
)

# ---------------- Utils (metrics / weights) ----------------
def _metrics_all(y_true, y_pred) -> Dict[str, float]:
    m = dict(
        ACC=accuracy_score(y_true, y_pred),
        BalACC=balanced_accuracy_score(y_true, y_pred),
        F1m=f1_score(y_true, y_pred, average="macro"),
        ARI=adjusted_rand_score(y_true, y_pred),
        V=v_measure_score(y_true, y_pred),
    )
    m["Score"] = 0.5*(m["ARI"] + m["V"])
    return m

def _make_sample_weights(y_like: np.ndarray) -> np.ndarray:
    vc = pd.Series(y_like).value_counts()
    K = float(vc.shape[0]); N = float(len(y_like))
    return np.array([(N/(K*(vc[y]+1e-9))) for y in y_like], dtype=np.float32)

# ---------------- Pseudobulk & ComBat ----------------
def _pseudobulk_sample_ct(adata: sc.AnnData, ct_col: str, sample_col: str) -> pd.DataFrame:
    obs = adata.obs[[ct_col, sample_col]].astype(str)
    groups = obs.groupby([sample_col, ct_col]).indices
    cols, names = [], []
    for (s, ct), idx in groups.items():
        X = adata.X[idx]
        cols.append(np.asarray(X.mean(axis=0)).ravel()); names.append(f"{s}|{ct}")
    return pd.DataFrame(np.column_stack(cols), index=adata.var_names.astype(str), columns=names)

def _combat(pb: pd.DataFrame) -> pd.DataFrame:
    ad = sc.AnnData(pb.T.copy())
    ad.obs["batch"] = pd.Categorical([c.split("|",1)[0] for c in pb.columns])  # Sample 為 batch
    sc.pp.combat(ad, key="batch")
    return pd.DataFrame(ad.X.T, index=pb.index, columns=pb.columns)

# ---------------- Marker selection (3 strategies) ----------------
def _per_ct_columns(pb: pd.DataFrame) -> Dict[str, List[str]]:
    cts = sorted({c.split("|",1)[1] for c in pb.columns})
    return {ct: [c for c in pb.columns if c.endswith("|"+ct)] for ct in cts}

def _logfc_mean(pb: pd.DataFrame, top_k: int) -> Dict[str, List[str]]:
    ct_cols = _per_ct_columns(pb)
    cts = list(ct_cols.keys())
    mean_all = pb.mean(axis=1) + 1e-9
    denom = max(1, len(cts)-1)
    out = {}
    for ct, cols in ct_cols.items():
        if not cols: out[ct]=[]; continue
        m_ct = pb[cols].mean(axis=1)
        rest = (mean_all * len(cts) - m_ct) / denom
        score = np.log2((m_ct+1e-9)/(rest+1e-9))
        out[ct] = list(score.sort_values(ascending=False).index[:top_k])
    return out

def _logfc_weighted_median(pb: pd.DataFrame, top_k: int, weights: pd.Series) -> Dict[str, List[str]]:
    # weights 索引需與 pb.columns 對齊（每個 Sample|CT 權重 = cell 數）
    ct_cols = _per_ct_columns(pb); cts = list(ct_cols.keys())
    mean_all = pb.mean(axis=1) + 1e-9
    denom = max(1, len(cts)-1)
    out = {}
    for ct, cols in ct_cols.items():
        if not cols: out[ct]=[]; continue
        w_ct = weights.reindex(cols).fillna(0.0).values
        W = w_ct / (w_ct.sum() + 1e-9)
        # 加權中位數：對每個基因，取樣本層的加權分位數 0.5
        X_ct = pb[cols].to_numpy()
        x_sorted = np.sort(X_ct, axis=1)
        # 近似：用加權均值替代加權中位數（快，對 heavy-tail 也穩）
        m_ct = (X_ct * W).sum(axis=1)
        m_ct = pd.Series(m_ct, index=pb.index)
        rest = (mean_all * len(cts) - m_ct) / denom
        score = np.log2((m_ct+1e-9)/(rest+1e-9))
        out[ct] = list(score.sort_values(ascending=False).index[:top_k])
    return out

def _bh_fdr(p: np.ndarray) -> np.ndarray:
    p = np.asarray(p, float); n = p.size
    order = np.argsort(p); ranked = p[order]
    q = ranked * n / (np.arange(n) + 1.0)
    q = np.minimum.accumulate(q[::-1])[::-1]
    out = np.empty_like(q); out[order] = q
    return out

def _logfc_mean_with_fdr(pb: pd.DataFrame, top_k: int) -> Dict[str, List[str]]:
    # 簡化：以樣本層 bootstrap 的兩樣本 t 近似 p 值（輕量版），再做 BH-FDR
    rng = np.random.default_rng(SEED)
    ct_cols = _per_ct_columns(pb); cts = list(ct_cols.keys())
    mean_all = pb.mean(axis=1) + 1e-9; denom = max(1, len(cts)-1)
    out = {}
    for ct, cols in ct_cols.items():
        if not cols: out[ct]=[]; continue
        m_ct = pb[cols].mean(axis=1)
        rest = (mean_all * len(cts) - m_ct) / denom
        lfc = np.log2((m_ct+1e-9)/(rest+1e-9)).to_numpy()

        # 以樣本列做簡易置換檢定估 p 值（迭代數小，保速度）
        X_ct = pb[cols].to_numpy()
        other_cols = [c for c in pb.columns if c not in cols]
        X_re = pb[other_cols].to_numpy()
        # 子抽 100 個對比（避免過慢）
        n_boot = min(100, X_re.shape[1])
        idx = rng.choice(X_re.shape[1], size=n_boot, replace=False)
        rest_boot = X_re[:, idx].mean(axis=1)
        lfc_boot = np.log2((m_ct+1e-9)/(rest_boot+1e-9)).to_numpy()
        # 近似 p：雙尾，假設置換分佈對稱
        diff = np.abs(lfc)
        ref  = np.abs(lfc_boot)
        # p ~ (# ref >= obs)/n_boot，避免 0
        pval = (np.sum(ref[None,:] >= diff[:,None], axis=1) + 1.0) / (n_boot + 1.0)
        fdr  = _bh_fdr(pval)

        keep = (fdr <= 0.05)
        # 先按顯著性過濾，不足再用 lfc 前排補齊到 top_k
        genes = pd.Series(lfc, index=pb.index)
        sel = list(genes[keep].sort_values(ascending=False).index[:top_k])
        if len(sel) < top_k:
            need = top_k - len(sel)
            extra = [g for g in genes.sort_values(ascending=False).index if g not in sel][:need]
            sel = sel + extra
        out[ct] = sel
    return out

def pick_markers(pb: pd.DataFrame, top_k: int, strategy: str, weights: pd.Series) -> Dict[str, List[str]]:
    if strategy == "logFC_mean":
        return _logfc_mean(pb, top_k)
    if strategy == "logFC_weighted_median":
        return _logfc_weighted_median(pb, top_k, weights)
    if strategy == "logFC_mean_fdr":
        return _logfc_mean_with_fdr(pb, top_k)
    raise ValueError(f"Unknown strategy: {strategy}")

# ---------------- Main: build markers -> CV with LGBM -> line plot ----------------
if __name__ == "__main__":
    ad = sc.read_h5ad(AD_PATH)
    for col in [LABEL_COL, SAMPLE_COL, CT_COL]:
        if col not in ad.obs: raise KeyError(f"obs 缺少欄位：{col}")
    # 輕量前處理（不篩細胞/基因）
    sc.pp.normalize_total(ad, target_sum=1e4)
    sc.pp.log1p(ad)

    # pseudo-bulk
    pb = _pseudobulk_sample_ct(ad, CT_COL, SAMPLE_COL)
    if USE_COMBAT:
        pb = _combat(pb)

    # 每個 Sample|CT 的 cell 數（作為權重）
    weights = (ad.obs[SAMPLE_COL].astype(str) + "|" + ad.obs[CT_COL].astype(str)).value_counts()
    weights = weights.reindex(pb.columns).fillna(0.0).astype(float)

    # y / groups（下游 CV 用）
    y_str = ad.obs[LABEL_COL].astype(str).values
    le = LabelEncoder().fit(y_str)
    y = le.transform(y_str)
    groups = ad.obs[SAMPLE_COL].astype(str).values
    K = len(le.classes_)
    sw_all = _make_sample_weights(y)

    results = []  # 收集 (strategy, topk, metrics)
    for strategy in STRATEGIES:
        print(f"\n=== Strategy: {strategy} ===")
        for topk in TOPK_LIST:
            # 1) 取各 CT Top-K
            ct_to_genes = pick_markers(pb, topk, strategy, weights)
            # 2) 取 union（去重）
            gene_union = sorted({g for gs in ct_to_genes.values() for g in gs if g in ad.var_names})
            if len(gene_union) == 0:
                raise ValueError(f"{strategy}/Top-{topk}: marker union 為空，請檢查資料。")

            # 3) 構建特徵矩陣（只取這些基因，直接用 ad.X；已正規化+log1p）
            ad_sub = ad[:, ad.var_names.isin(gene_union)].copy()
            X = np.asarray(ad_sub.X.todense() if hasattr(ad_sub.X, "todense") else ad_sub.X, dtype=np.float32)
            if X.ndim != 2: raise ValueError("X 維度錯誤")

            # 4) GroupKFold CV
            gkf = GroupKFold(n_splits=min(N_SPLITS, np.unique(groups).size))
            fold_rows = []
            print(f"[Top-{topk}] #genes={X.shape[1]}")
            for fold, (tr_idx, va_idx) in enumerate(gkf.split(X, y, groups), 1):
                val_groups = sorted(set(groups[va_idx].tolist()))
                print(f"  - Fold {fold}: val_groups={val_groups} (n_val={len(va_idx)})")

                X_tr, X_va = X[tr_idx], X[va_idx]
                y_tr, y_va = y[tr_idx], y[va_idx]
                sw_tr = sw_all[tr_idx]

                clf = lgb.LGBMClassifier(num_class=K, **LGB_PARAMS)
                clf.fit(
                    X_tr, y_tr,
                    sample_weight=sw_tr,
                    eval_set=[(X_va, y_va)],
                    eval_metric="multi_logloss",
                    callbacks=[lgb.early_stopping(stopping_rounds=EARLY_STOP, verbose=False)],
                )
                best_it = clf.best_iteration_ or LGB_PARAMS["n_estimators"]
                yhat = clf.predict(X_va, num_iteration=best_it)
                m = _metrics_all(y_va, yhat)
                m.update({"strategy": strategy, "TopK": int(topk), "fold": fold})
                fold_rows.append(m)

            df_cv = pd.DataFrame(fold_rows)
            mean_row = {"strategy": strategy, "TopK": int(topk), "fold": "mean"}
            for c in ["ACC","BalACC","F1m","ARI","V","Score"]:
                mean_row[c] = float(df_cv[c].mean())
            results.append(mean_row)
            print(f"  => {strategy} / Top-{topk}: F1m={mean_row['F1m']:.4f} | BalACC={mean_row['BalACC']:.4f} | Score={mean_row['Score']:.4f}")

    df_out = pd.DataFrame(results).sort_values(["strategy","TopK"]).reset_index(drop=True)
    print("\n[Summary] Biomarker-only sweep (LGBM, GroupKFold)")
    print(df_out.round(4).to_string(index=False))

    # 折線圖：不同策略各一條
    plt.figure(figsize=(8,5))
    for strategy in STRATEGIES:
        sub = df_out[df_out["strategy"]==strategy]
        plt.plot(sub["TopK"], sub["F1m"], marker="o", label=f"{strategy} — F1m")
    plt.xlabel("Top-K per CT (marker-only)")
    plt.ylabel("CV F1-macro")
    plt.title("Biomarker-only: selection strategy × Top-K (LGBM, ES=50, GroupKFold)")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
