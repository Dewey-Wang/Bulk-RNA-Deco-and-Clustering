```python
# Jupyter cell: zip existing CSVs in results_path
from pathlib import Path
import zipfile
from datetime import datetime

# --- settings ---
results_path = Path("./result_final")   # 保持你的目錄
archive_name = "Wang_Ding-Yang_Project2.zip"
files_to_pack = ["pred_props.csv", "cluster_membership.csv"]
OVERWRITE = True                             # True: 直接覆寫同名 zip；False: 自動加時間戳

# --- resolve archive path (optionally timestamp to avoid overwrite) ---
zip_path = results_path / archive_name
if zip_path.exists():
    zip_path.unlink()  # 刪除舊的 zip


# --- sanity checks ---
missing = [f for f in files_to_pack if not (results_path / f).exists()]
if missing:
    raise FileNotFoundError(f"下列檔案不存在於 {results_path}: {missing}")

# --- create zip with deflate compression ---
with zipfile.ZipFile(zip_path, mode="w", compression=zipfile.ZIP_DEFLATED) as zf:
    for fname in files_to_pack:
        fp = results_path / fname
        # arcname 只放檔名，避免把整個路徑打到 zip 裡
        zf.write(fp, arcname=fname)

print(f"[OK] Wrote: {zip_path} | files: {files_to_pack}")
```
